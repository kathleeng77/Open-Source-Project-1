```python
%matplotlib notebook
import matplotlib.pyplot as plt
import numpy as np
import pandas as pd
```

```python
birth_rate_data = pd.read_csv("data/crude_birth_rate.csv")
birth_rate_data
```

```python
x_axis = birth_rate_data["DATE"]
y_axis = birth_rate_data["BIRTHS PER 1,000 PEOPLE"]

plt.xlabel("Years")
plt.ylabel("Birth Rate per 1,000 People")
plt.title ("Crude Birth Rate in US (1960 - 2017)")

plt.plot(x_axis, y_axis)

plt.savefig("Crude Birth Rates in US.png")

plt.show()
```
