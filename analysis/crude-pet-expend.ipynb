```python
%matplotlib notebook
import matplotlib.pyplot as plt
import numpy as np
import pandas as pd
```

```python
pet_expend_data = pd.read_csv("data/crude_pet_expend.csv")
pet_expend_data
```

```python
x_axis = pet_expend_data["DATE"]
y_axis = pet_expend_data["PET EXPENDITURES IN BILLIONS"]

plt.xlabel("Years")
plt.ylabel("Pet Expenditures (in Billions)")
plt.title ("Pet Expenditure Trend in US (1960 - 2017)")

plt.plot(x_axis, y_axis)

plt.savefig("Pet Expend.png")

plt.show()
```
